```python
def mean_attr(X,Y,Z): # U1 : Mean Attribute (10%)
    a = X[Z][0]
    try:
            a == int(a) ## 만약 숫자를 입력하지 않는  다면 예외처리!
    except ValueError:
            print("Input number value!")
            pass
            return
    meanDiff=X[Z].mean()-Y[Z].mean()
    return abs(meanDiff)
```

```python
def mean_corr(X,Y,Z): # U2 : Mean Correlation (10%)
    mean_sum = 0
    for i in Z:
        a = X[i][0] ## 데이터 하나를 추출하여 int형으로 강제 형변환 한것과 비교,,, 같은 경우 정수형으로 보고 인식
        try:
                a == int(a) ## 만약 숫자를 입력하지 않는  다면 예외처리!
        except ValueError: ## 값이 일치하지 않을 경우
                print("Input number value!") ## 에러메세지 출력
                pass
                return ## NULL 값 반환(아무것도 반환 안함)
        df_corr = pd.DataFrame({'x': X[i], 'y': Y[i]}) ## KL과정을 거치기 전 df X와 거친 후 Y의 Z열(KL처리를 한 항목) 값을 받아서 데이터프레임 생성
        meanCorr=df_corr.corr(method='pearson') ## 피어슨 상관계수 계산 (df 자체적으로 돌아감)
        print(df_corr.corr(method='pearson'))
        print(meanCorr.iloc[0,1])
        mean_sum += meanCorr.iloc[0,1]
    return abs(mean_sum) ## 피어슨 상관계수 값 추출 0,1 즉 서로 다른 열을 계산한 값을 출력함,,,, 아래의 출력창 참고
```

```python
k = {'구매액','수량','전년도 누적금액'}
print(mean_corr(df,df3,k))
```

```python
def cos_sim(X, Y, Z): # U3 : Cosine Similarity (10%)
    a = X[Z][0]
    try:
            a == int(a) ## 만약 숫자를 입력하지 않는  다면 예외처리!
    except ValueError:
            print("Input number value!")
            pass
            return
    meanSim = dot(X[Z],Y[Z])/(norm(X[Z])*norm(Y[Z]))    
    return meanSim
```
