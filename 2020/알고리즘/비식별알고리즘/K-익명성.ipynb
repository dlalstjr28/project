```python
import random
import pandas as pd

df = pd.read_csv("test.csv")
df
```

```python
def K_anonymity(df, number): 
        try:
            number = int(number) ## 만약 숫자를 입력하지 않는  다면 예외처리!
        except NameError:
            print("Input number value!")
            pass
        
        list = ['주소','나이','성별'] ## 우리가 처리할 준식별정보
        ## list 는 준식별 정보를 담기 위해 for 문 사용 권장
        #print(df.groupby(list)[list[0]].count())
        #df['count'] = df.groupby(list)[list[0]].count()
        df['count'] = df.groupby(list)[list[0]].transform('size') ## list[0]으로 최소 칼럼 보장.
        #df['count'] = df.groupby(by=list, as_index=False).count() ## list[0]으로 최소 칼럼 보장.j
        #print(df.groupby(by=list, as_index=False).count())
        ##list의 데이터를 그룹화 한 후 list[0]으로 최소 집계해서 size를 구해 count에 결과 저장
        df = df.loc[df['count']>=number] 
        #number(k)보다 높거나 같은 size만 df에서 추출
        df_index = df.index ## 인덱스가 초기화 되므로 일련번호 저장
        ##일련 번호(맨 앞 0,1,2,3, 순서) 새로 초기화
        #print(df['count'])
        del df['count']
        df = df.set_index(df_index)
        return df
```
