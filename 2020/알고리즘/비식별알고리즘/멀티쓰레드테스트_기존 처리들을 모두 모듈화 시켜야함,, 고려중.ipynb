```python
import threading,time
```

```python
def test1():
    for i in range(5):

        time.sleep(1)

        print("{0}sec".format(i))
    
    return 1

def test2():

    for i in range(3):

        time.sleep(1)

        print("{0}second".format(i))
    
    return 1

def test3():
    
    for i in range(7):

        time.sleep(1)

        print("{0}초 지남".format(i))
        
    return 1

list = [test1, test2, test3]
```

```python
x = 2
t1=threading.Thread(target=list[0])
t1.start()
print("1 번 동작")
t2=threading.Thread(target=list[1])
t2.start()
print("2 번 동작")
while x < 12:
    test_select = x%3
    if not t1.is_alive():
        t1=threading.Thread(target=list[test_select])
        t1.start()
        x += 1
        print(x,"번 동작")
        continue
    if not t2.is_alive():
        t2=threading.Thread(target=list[test_select])
        t2.start()
        x += 1
        print(x,"번 동작")
        continue
```
