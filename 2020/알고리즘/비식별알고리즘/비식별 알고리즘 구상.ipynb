```python
# 나이 생성. 나이대별 퍼센트별로 랜덤 생성인듯?
def randOld():
  rv = norm(loc = 44 , scale = 10)
  y = rv.rvs(size=1)
  return int(y) ## 65세 이상 13.8 퍼센트 /14세 이상 65세 미만 73.2퍼센트 / 14세 미만 13.1퍼센트
```

```python
# 가족관계 생성. 랜덤 추출
def randfamily(input):
    family = ['외동딸', '외동아들', '1남1녀', '1남2녀', '2남1녀', '다자녀', '실부실모', '편부', '편모']
    if input["sex"] == "F" :     # 여자면 외동아들 제외
        family.remove("외동아들")
    else :                       # 남자면 외동딸 제외
        family.remove("외동딸")
    return random.choice(family)
```

```python
# 예시 데이터 생성

data_count = 100      # 생성할 데이터 개수

fake = Faker("ko_KR")

df = pd.DataFrame(columns=('name',
                           'email',
                           'phone',
                           'age',
                           'sex',
                           'address',
                           # 'weight' 아래에서 추가
                           # 'Port' 아래에서 추가
                           # 'Credit' 아래에서 추가
                           'family'
                           # 'Operation' 아래에서 추가
                          )
                 )

for i in range(data_count):
    buffer = fake.profile()
    stuff = [buffer["name"],
             buffer["mail"],
             randPhoneNumber(),
             randOld(),
             buffer["sex"],
             randAddress(),
             # weight 아래에서 추가
             # port 아래에서 추가
             # credit 아래에서 추가
             randfamily(buffer)
             # operation 아래에서 추가
            ]
    df.loc[i] = [item for item in stuff]

df = df.astype({'age':'int'})
```

```python
# 몸무게 추가
def randWeight(df):
    weightList=[] # df 항목들의 weight가 쌓일 list
    
    for i in df.index:
        val = df.loc[i] # for문이 실행되는 row 
        weight = 0 # weight 변수 초기화

        if val.sex == "M": # 남성일 때
            if val.age < 30: # sizeofkorea를 기반으로 나이에 따라 대략적인 평균몸무게 산출
                weight = 13/14*val.age+47.8
            else: # val.age >= 30:
                weight = -3/25*(val.age-30)+76.3

        else: # val.sex == "F" 여성일 때
            if val.age < 20:
                weight = 7/4*val.age+91/4
            elif val.age < 40:
                weight = 4/15*val.age+(149/3)
            else: # val.age >= 40:
                weight = 0.1*val.age+54

        weightList.append(format(weight,".2f")) # 산출된 몸무게를 list에 append

    df.loc[:,'weight'] = weightList # for문을 통해 산출된 몸무게 리스트를 사용해 "weight" column에 추가
    
    return df
randWeight(df)
```
