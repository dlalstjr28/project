```python
# 예시 데이터 생성

import pandas as pd
import random
from faker import Faker
fake = Faker("ko_KR")
import numpy as np


df = pd.DataFrame(columns=('name'
    , 'email'
    , 'phone'
    , 'age'
    , 'sex'
    , 'address'
    , 'city'))

for i in range(10):
    stuff = [fake.name()
        , fake.email()
        , fake.phone_number()
        , fake.pyint(min_value=5, max_value=80)
        , random.choice("FM")
        , fake.address()
        , fake.city()]
    df.loc[i] = [item for item in stuff]

df = df.astype({'age':'int'})
```

```python
import hashlib
class anonymous() :
    def __init__(self,df):
        self.df = df
        self.backup = backup
        
    def SHA256withSalt(self, columns, salt):
        """
            SHA256withSalt : sha256 해시 함수 + salt
        """
        self.backup = self[columns]
        SHA256 = hashlib.sha256() # sha256 hash 객체 생성
        col = []
        for i in range(len(df.index)):
            SHA256.update((str(df.loc[i][columns])+str(salt)).encode('euc-kr'))
            col.append(SHA256.hexdigest()) # sha 256 해시에 대한 16진수 값 산출
            #df[columns, i] = SHA256.hexdigest()
        df[columns]= col
        #new_col = np.array(col)
        return df
    
    def delete_col(self, columns):
        """
            delete_col : 열 삭제
            columns : 삭제할 column명
        """
        self.backup = self[columns]
        del_df = df.drop(columns, axis=1)
        return del_df
    
    def delete_row(self, rows):
        """
            delete_row : 행 삭제
            - rows : 삭제할 row명
        """
        self.backup = self.iloc[rows]
        del_df = df.drop(rows, axis=0)
        return del_df
    
    def delete_partial(self, column, num):
        """
            delete_partial : 부분 삭제
            - column : 부분 삭제할 column명
            - num : num번째 이후 값 삭제
        """
        self.backup = self[column]
        data = df[column].str.split(' ')
        col = []
        for i in data:
            i = i[:num]
            col.append(" ".join(i))
        return col
```
